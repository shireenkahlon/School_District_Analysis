```python
#make a list of prefixes and suffixes
prefixes_suffixes = []
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
#iterate through prefixes_suffices to replace the prefix and suffix
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")
student_data_df
```

```python
student_data_df.tail(50)
```

```python
#merge tables together
student_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
student_data_complete_df
```

```python
#count the number of values
# student_count = student_data_complete_df.count()
# student_count
#count the number per column
student_count = student_data_complete_df["Student ID"].count()
student_count
```

```python
school_count = len(student_data_complete_df["school_name"].unique())
school_count
```

```python
#get the budget but make sure to tag original school data table
school_budget = school_data_df["budget"].sum()
school_budget
```

```python
#find the average of reading scores
reading_average = student_data_complete_df["reading_score"].mean()
reading_average
```

```python
#find the average of math scores
math_average = student_data_complete_df["math_score"].mean()
math_average
```

```python
#find the percentage of reading and math
passing_reading = student_data_complete_df[student_data_complete_df["reading_score"] >= 70]
passing_math = student_data_complete_df[student_data_complete_df["math_score"] >= 70]
#calculate the number of students who passed reading
passing_reading_count = passing_reading["student_name"].count()
#calculate the number of students who passed math
passing_math_count = passing_math["student_name"].count()
print(passing_reading_count)
print(passing_math_count)
```

```python
#calculate reading and math percentages
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_math_percentage = passing_math_count / float(student_count) * 100
print(passing_reading_percentage)
print(passing_math_percentage)
```

```python
#find both the passing math and reading scores
passing_math_reading = student_data_complete_df[(student_data_complete_df["reading_score"] >= 70) & (student_data_complete_df["math_score"] >= 70)]
passing_math_reading
```

```python
#count the number of students who passed both math and reading
overall_passing_count = passing_math_reading["student_name"].count()
overall_passing_count
```

```python
#find the percentage of students who passed both math and reading
overall_passing_percentage = overall_passing_count / float(student_count) * 100
overall_passing_percentage
```

```python
district_summary_df = pd.DataFrame(
        [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": school_budget,
          "Average Math Score": math_average,
          "Average Reading Score": reading_average,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]


```

```python
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Budget"]
                                                                            
```

```python
# Format the columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)

district_summary_df
```
